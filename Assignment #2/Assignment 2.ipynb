```python
speed_mean=int(data["Speed"].mean())

def set_speed(val):
    if val<speed_mean:
        return "Speed Low"
    elif val==speed_mean:
        return "Speed neutral"
    else: 
        return "Speed High"

data["High_Low_Speed"]=data["Speed"].apply(set_speed)
data
```

```python
HP_mean=int(data["HP"].mean())

def set_hp(val):
    if val<HP_mean:
        return "HP Low"
  
    else: 
        return "HP High"

data["High_Low_HP"]=data["HP"].apply(set_hp)
data
```
